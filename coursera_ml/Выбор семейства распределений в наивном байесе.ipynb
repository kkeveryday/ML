```
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.datasets import load_breast_cancer
```

```
digits_X, digits_y = load_digits(return_X_y=True)
breast_cancer_X, breast_cancer_y = load_breast_cancer(return_X_y=True)
```

```
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB

#Digits
bnb = BernoulliNB()
print('cross val score BerniulliNB/digits =', cross_val_score(bnb, digits_X, digits_y).mean())
mnb = MultinomialNB()
print('cross val score MultinomialNB/digits =', cross_val_score(mnb, digits_X, digits_y).mean())
gnb = GaussianNB()
print('cross val score GaussianNB/digits =', cross_val_score(gnb, digits_X, digits_y).mean())
print('----------------------------------------------------------------')
#Breast cancer
bnb = BernoulliNB()
print('cross val score BerniulliNB/breast_cancer =', cross_val_score(bnb, breast_cancer_X, breast_cancer_y).mean())
mnb = MultinomialNB()
print('cross val score MultinomialNB/breast_cancer =', cross_val_score(mnb, breast_cancer_X, breast_cancer_y).mean())
gnb = GaussianNB()
print('cross val score GaussianNB/breast_cancer =', cross_val_score(gnb, breast_cancer_X, breast_cancer_y).mean())
```
