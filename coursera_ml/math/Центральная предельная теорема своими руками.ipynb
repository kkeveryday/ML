#[Распределение Лапласа](https://ru.wikipedia.org/wiki/Распределение_Лапласа)


---


Распределе́ние Лапла́са (двойно́е экспоненциа́льное) — в теории вероятностей это непрерывное распределение случайной величины, при котором плотность вероятности есть $f(x) = \frac{\alpha}{2} e^{-\alpha|x-\beta|}, -\infty < x < +\infty$, где $\alpha > 0$ - параметр масшатаба, $-\infty < \beta < +\infty$ - параметр сдвига.

---
По определению, функция распределения — это интеграл от плотности распределения: $F(x) = \int_{-\infty}^x f(t)dt = \frac{\alpha}{2}\int_{-\infty}^x e^{-\alpha|x-\beta|}dt$.




```
# Рассматривать распределение будем при apha = 1 и beta = 0
def laplace_pdf(x):
  return 0.5 * np.exp(-abs(x))


# Функция генерации выборки по лапласу 
def rvs(size=None, random_state=None):
  return np.random.laplace(0, 1, size=size)


# Функция нормальнального распределения с параметрами
def norm_pdf(mu, sigma, x):
  return np.exp(-0.5 * ((x - mu) / sigma) ** 2) / sigma / np.sqrt(2 * np.pi)
```

## Оценка распределения выборочного среднего при n = 5

```
n = 5
sample_5 = []
for ш in range(1000):
    sample_temp = rvs(n) # Делаем выборку из 5
    mean = sum(sample_temp) / n # Определяем среднее
    sample_5.append(mean)
norm_x = np.linspace(-3, 3, 1000)
var_norm = var / n # Определяем дисперсию для нормального распределения
sigma_norm = var_norm ** 0.5 # Находим сигму
norm_5_pdf = norm_pdf(mu=mu, sigma=sigma_norm, x=norm_x) # Строим ДФР
```

```
# Построение графиков
plt.plot(norm_x, norm_5_pdf, label='Нормальное распределение');
plt.hist(sample_5, bins=100, density=True, label='Гистограмма средних');
plt.legend();
```
