```
import re
import numpy as np
from scipy.spatial import distance
```

```
with open('sentences.txt', 'r') as file:
  text = file.read()
```

```
text = text.lower()
print(text)
```

```
all_words = re.split('[^a-z]', text)
print(len(all_words))
```

```
words = [word for word in all_words if word != '']
len(words)
```

```
unique_words = set(words)
words_dict = {}
i = 0
for word in unique_words:
  words_dict[i] = word
  i += 1
len(words_dict)
```

```
text = text.split('\n')
text = [sentence for sentence in text if sentence != '']
len(text)
```

```
sentences_list = []
for sentence in text:
  sentence_split = re.split('[^a-z]', sentence)
  sentence_split = [word for word in sentence_split if word != '']
  sentences_list.append(sentence_split)
```

```
matrix = np.ones((22, 254), dtype=int)
for i in range(len(text)):
  for j in range(len(words_dict)):
    matrix[i][j] = sentences_list[i].count(words_dict[j])
```

```
distance_dict = {}
for i in range(1, len(text)):
  distance_dict[i] = distance.cosine(matrix[0], matrix[i])
```

Предложения под номером 6 и 4 больше всего совпадают в предложением под номером 0

```
list_distance_dict = list(distance_dict.items())
list_distance_dict.sort(key=lambda i: i[1])
list_distance_dict
```
