```
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 8
```

```
def f(x):
  return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```
dots = range(1, 16)
func = []
for dot in dots:
  func.append(f(dot))
plt.plot(func);
```

#1я итерация для точек 1 и 15

```
A = np.ones((2, 2), dtype=int)
A[0][1] = 1
A[1][1] = 15
b = [f(1), f(15)]
```

```
w = linalg.solve(A, b)
w
```

```
def f1(x):
  return w[0] + w[1] * x
```

```
func1 = []
for dot in dots:
  func1.append(f1(dot))
plt.plot(func1);
```

#2я итерация для точек 1, 8 и 15

```
A = np.ones((3, 3), dtype=int)
A[0][1] = 1
A[1][1] = 8
A[2][1] = 15
A[0][2] = 1
A[1][2] = 64
A[2][2] = 225
b = [f(1),f(8), f(15)]
```

```
w = linalg.solve(A, b)
w
```

```
def f2(x):
  return w[0] + w[1] * x + w[2] * x * x
```

```
func2 = []
for dot in dots:
  func2.append(f2(dot))
plt.plot(func2);
```

#3я итерация для точек 1, 4, 10 и 15

```
A = np.ones((4, 4), dtype=int)
A[0][1] = 1
A[1][1] = 4
A[2][1] = 10
A[3][1] = 15
A[0][2] = 1
A[1][2] = 16
A[2][2] = 100
A[3][2] = 225
A[0][3] = 1
A[1][3] = 64
A[2][3] = 1000
A[3][3] = 3375
b = [f(1),f(4),f(10), f(15)]
```

```
w = linalg.solve(A, b)
w
```

```
def f3(x):
  return w[0] + w[1] * x + w[2] * x * x + w[3] * x * x * x
```

```
func3 = []
for dot in dots:
  func3.append(f3(dot))
plt.plot(func3);
```
