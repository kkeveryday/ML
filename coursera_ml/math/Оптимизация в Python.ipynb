```
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 8
```

#Минимизация гладкой функции
Функция f(x)

```
dots = np.linspace(1, 30, 30, dtype=int)
def f(x):
  return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```
func = []
for dot in dots:
  func.append(f(dot))
plt.plot(dots, func)
```

```
from scipy.optimize import minimize

res1 = minimize(f, 4, method='BFGS')
res2 = minimize(f, 30, method='BFGS')
print('Значение функции:', res1.fun)
print('Значение функции:', res2.fun)
```

#Глобальная оптимизация
Функция f(x)

```
from scipy.optimize import differential_evolution

res = differential_evolution(f, [(1, 30)])
print('Значение функции:', res.fun)
```

#Минимизация негладкой функции
Функция h(x) = int(f(x))

```
def h(x):
  return int(np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2))
```

```
func1 = []
for dot in dots:
  func1.append(h(dot))
plt.plot(dots, func1)
```

```
res1 = minimize(h, 30, method='BFGS')
print('Значение функции:', res1.fun)
res2 = differential_evolution(h, [(1, 30)])
print('Значение функции:', res2.fun)
```
