```
import numpy as np
from sklearn.datasets import load_digits
```

```
from sklearn.model_selection import train_test_split

X, y = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
idx = int(0.75 * X.shape[0]) + 1
X_train = X[:idx]
X_test = X[idx:]
y_train = y[:idx]
y_test = y[idx:]
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knc = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knc.fit(X_train, y_train)
print('knn classifier error =', 1 - accuracy_score(y_test, knc.predict(X_test)))
```

```
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train, y_train)
print('rfc classifier error =', 1 - accuracy_score(y_test, rfc.predict(X_test)))
```
